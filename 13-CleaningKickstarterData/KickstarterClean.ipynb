```python
import pandas as pd
```

```python
path = "Resources/KickstarterData.csv"
kickstarter_df = pd.read_csv(path)
```

```python
kickstarter_df.head()
```

```python
#extracting relevant columns for analysis
extracted_kickstarter_df = kickstarter_df[["name", "goal", "pledged", "state", "country", "staff_pick",
                                          "backers_count", "spotlight"]]
extracted_kickstarter_df.head()
```

```python
# removing projects that had no pledges
extracted_kickstarter_df = extracted_kickstarter_df.loc[extracted_kickstarter_df["pledged"]>0]
extracted_kickstarter_df.head()
```

```python
#creating new dataframe that is filtered to campaigns only in the US
columns = ["name", "goal", "pledged", "state", "country","staff_pick", "backers_count", "spotlight"]US
usa_kickstarter_df = extracted_kickstarter_df.loc[extracted_kickstarter_df["country"]=="US", columns]
usa_kickstarter_df.head()
```

```python
# column that has average amount pledged per backer
average_pledge = usa_kickstarter_df["pledged"]/usa_kickstarter_df["backers_count"]
average_pledge
```

```python
#formating Goals and Pledge columns and also adding in average donation with same formatting
usa_kickstarter_df["goal"] = pd.to_numeric(usa_kickstarter_df["goal"])
usa_kickstarter_df["pledged"] = pd.to_numeric(usa_kickstarter_df["pledged"])
usa_kickstarter_df["goal"] = usa_kickstarter_df["goal"].map("${:,.2f}".format)
usa_kickstarter_df["pledged"] = usa_kickstarter_df["pledged"].map("${:,.2f}".format)
usa_kickstarter_df["average donation"] = average_pledge.map("${:,.2f}".format)
usa_kickstarter_df.head()
```

```python
#total number of backers in the data
total_backers = usa_kickstarter_df["backers_count"].sum()
total_backers
```

```python
#average number of backers for US campaigns
average_backers = usa_kickstarter_df["backers_count"].mean()
average_backers
```

```python
#creating new df for only those campaigns that were staff picked
staffpicked_kickstarter_df = usa_kickstarter_df.loc[usa_kickstarter_df["staff_pick"]==True]
staffpicked_kickstarter_df.head()
```

```python
# grouping by the current state of each staff-picked campaign (about 83%)
grouped_staffpicked_df = staffpicked_kickstarter_df.groupby(['state'])
grouped_staffpicked_df["name"].count()
```
