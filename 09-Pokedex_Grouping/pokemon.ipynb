```python
import pandas as pd
```

```python
path = "Resources/Pokemon.csv"
```

```python
pokemon_df = pd.read_csv(path)
pokemon_df.head()
```

```python
pokemon_stats_df = pokemon_df[["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_stats_df.head()
```

```python
grouped_stats_df = pokemon_stats_df.groupby(["Type 1"])
grouped_stats_df.mean()
```

```python
pokemon_df["Total Power Level"] = pokemon_stats_df[["HP", "Attack", "Sp. Atk","Sp. Def", "Speed"]].sum(axis=1)
pokemon_df.head()
```

```python
strongest_type_df = pokemon_df[["Type 1", "Total Power Level"]]
grouped_strongest_df = strongest_type_df.groupby(["Type 1"])
sum_grouped_strongest = grouped_strongest_df.sum()
sum_grouped_strongest
```

```python
sum_grouped_strongest.sort_values(by=["Total Power Level"], inplace=True, ascending=False)
sum_grouped_strongest
```

```python
sum_grouped_strongest.to_csv("output/TypeRanks.csv", index=True)
```
