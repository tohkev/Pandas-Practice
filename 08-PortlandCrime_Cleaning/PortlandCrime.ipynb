```python
#Importing dependencies
import pandas as pd
```

```python
#reference path where CSV file is located
path = "Resources/crime_incident_data2017.csv"
```

```python
#reading and converting to DataFrame
incident_df = pd.read_csv(path)
incident_df.head()
```

```python
#looking for missing values
incident_df.count()
```

```python
#removing rows with null values
incident_df = incident_df.dropna(how="any")
incident_df.head()
```

```python
#verifying counts
incident_df.count()
```

```python
#checking to see if there are any similar or mispelled values
incident_df["Offense Type"].value_counts()
```

```python
#combining similar offenses together
incident_df = incident_df.replace({"Theft From Motor Vehicle":"Motor Vehicle Theft"})
```

```python
incident_df["Offense Type"].value_counts()
```

```python
incident_df = incident_df.replace({"Theft from Building":"Robbery", "Purse-Snatching":"Robbery", "Stolen Property Offenses":"Robbery", "Commercial Sex Acts":"Prostitution", "Assisting or Promoting Prostitution":"Prostitution"})
incident_df["Offense Type"].value_counts()
```

```python
#creating a new DataFrame that looks into King neighborhood
king_incidents = incident_df.loc[incident_df["Neighborhood"]=="King",:]
king_incidents.head()
```
