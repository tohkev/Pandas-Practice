```python
import pandas as pd
```

```python
bitcoin_path = "Resources/bitcoin_cash_price.csv"
dash_path = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_path)
bitcoin_df.head()
```

```python
dash_df = pd.read_csv(dash_path)
dash_df.head()
```

```python
merged_df = pd.merge(bitcoin_df, dash_df, on="Date")
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={
    "Open_x":"Bitcoin Open",
    "High_x":"Bitcoin High",
    "Low_x":"Bitcoin Low",
    "Close_x":"Bitcoin Close",
    "Volume_x":"Bitcoin Volume",
    "Market Cap_x":"Bitcoin Market Cap",
    "Open_y":"Dash Open",
    "High_y":"Dash High",
    "Low_y":"Dash Low",
    "Close_y":"Dash Close",
    "Volume_y":"Dash Volume",
    "Market Cap_y":"Dash Market Cap"
})
merged_df.head()
```

```python
max_btc_open = merged_df["Bitcoin Open"].max()
max_dash_open = merged_df["Dash Open"].max()
```

```python
max_btc_close = merged_df["Bitcoin Close"].max()
max_dash_close = merged_df["Dash Close"].max()
```

```python
total_btc_vol = round(merged_df["Bitcoin Volume"].sum()/1000000,2)
total_dash_vol = round(merged_df["Dash Volume"].sum()/1000000,2)
```

```python
summary_dict = {
    "Best BTC Open":max_btc_open,
    "Best Dash Open":max_dash_open,
    "Best BTC Close":max_btc_close,
    "Best Dash Close":max_dash_close,
    "Total BTC Volume (in millions)":total_btc_vol,
    "Total Dash Volume (in millions)":total_dash_vol
}
summary_table = pd.DataFrame(summary_dict, index=[0])
summary_table
```
