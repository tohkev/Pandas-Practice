```python
#importing dependencies
import pandas as pd
```

```python
path = "Resources/ted_talks.csv"
```

```python
tedtalk_df = pd.read_csv(path)
tedtalk_df.head()
```

```python
#getting ranges for max and min values 
min_views = tedtalk_df["views"].min()
min_views
```

```python
max_views = tedtalk_df["views"].max()
max_views
```

```python
#creating bins based on max and min values
bins = [0, 199999, 399999, 599999, 799999, 999999, 1999999, 2999999, 3999999, 4999999, 50000000]
group_labels = ["0 to 199k", "200k to 399k", "400k to 599k", "600k to 799k","800k to 999k","1mil to 2mil", "2mil to 3mil", "3mil to 4mil","4mil to 5mil", "5mil to 50mil"]
```

```python
#slicing data and distributing to bins
tedtalk_df["View Group"] = pd.cut(tedtalk_df["views"], bins, labels=group_labels)
tedtalk_df.head()
```

```python
#creating groupby object based on the newly created column
tedtalk_group_df = tedtalk_df.groupby(["View Group"])
tedtalk_group_df.head()
```

```python
#getting the number of rows in each bin
tedtalk_group_df["View Group"].value_counts()
```

```python
#finding average amount of comments based on each bin
average_comments = tedtalk_group_df["comments"].mean()
average_comments
```

```python
#finding average amount duration based on each bin
average_duration = tedtalk_group_df["duration"].mean()
average_duration
```

```python
#finding average amount of languages based on each bin
average_languages = tedtalk_group_df["languages"].mean()
average_languages
```
